# Merge Stations

```python
# ADD STATION AND LOCATION
#Display catalog of stations 
#The data frame has information about the station location in coordinates, city, and township
filen="./Resources/cat_estacion.csv"
df2 = pd.read_csv(filen, encoding="ISO-8859-1")
df2.head()
```

```python
#Merge
for index, row in df2.iterrows():
    s_lon = row['longitud']
    s_lat = row['latitud']
    s_station = row['cve_estac']
    s_township = row['Municipio']
    df.loc[df['Station'] ==s_station, 'Lat'] = s_lat
    df.loc[df['Station'] ==s_station, 'Long'] = s_lon
    df.loc[df['Station'] ==s_station, 'Municipality'] = s_township
```

```python
print(df.shape)
df.head()
```
